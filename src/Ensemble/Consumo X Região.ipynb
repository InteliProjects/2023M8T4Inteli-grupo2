```python
# Função para remover acentos
def remove_accent(text):
    return unidecode(str(text))

# Aplicar a função para remover acentos nas colunas desejadas
columns_to_remove_accent = ['Abrangencia', 'Regiao']  # Substitua 'coluna1' e 'coluna2' pelos nomes reais das colunas
for coluna in columns_to_remove_accent:
    cep[coluna] = cep[coluna].apply(remove_accent)
```

```python
cnpj = pd.read_csv('./CNPJ/cnpj_3_s3.csv', encoding='latin1', sep=',')
```

```python
cnpj.info()
```

```python
# Selecione as primeiras 5 linhas do DataFrame cnpj
ceps = cnpj.head(40000).copy()

cep_amostra = cep.sample(n=min(73, cep.shape[0]), replace=False, random_state=42)


# Selecione as colunas relevantes nos DataFrames
cep_regioes = cep[['Regiao', 'Faixa_CEP']]
ceps = ceps[['cep', 'cnpj']].copy()

# Converta os códigos de CEP para string para garantir a compatibilidade
cep_regioes['Faixa_CEP'] = cep_regioes['Faixa_CEP'].astype(str)
ceps['cep'] = ceps['cep'].astype(str)

# Separe os intervalos de CEP em duas colunas: Faixa_CEP_Inicial e Faixa_CEP_Final
cep_regioes[['Faixa_CEP_Inicial', 'Faixa_CEP_Final']] = cep_regioes['Faixa_CEP'].str.split(' - ', expand=True)

# Converta as colunas 'Faixa_CEP_Inicial' e 'Faixa_CEP_Final' para tipo inteiro
cep_regioes['Faixa_CEP_Inicial'] = cep_regioes['Faixa_CEP_Inicial'].astype(int)
cep_regioes['Faixa_CEP_Final'] = cep_regioes['Faixa_CEP_Final'].astype(int)
ceps['cep'] = ceps['cep'].astype(int)

# Criar um DataFrame vazio para armazenar os resultados
resultado = pd.DataFrame(columns=['cep', 'cnpj', 'Faixa_CEP', 'Regiao'])

# Iterar pelas linhas do DataFrame ceps
for index, row in ceps.iterrows():
    cep_atual = str(row['cep'])[:5]  # Pegar os primeiros cinco dígitos do CEP

    # Encontrar a faixa correspondente no DataFrame cep
    faixa_correspondente = cep_regioes[
        (cep_regioes['Faixa_CEP_Inicial'] <= int(cep_atual)) &
        (cep_regioes['Faixa_CEP_Final'] >= int(cep_atual))
    ]

    # Se encontrar uma faixa correspondente, adicionar ao DataFrame resultado
    if not faixa_correspondente.empty:
        resultado = resultado.append({
            'cep': row['cep'],
            'cnpj': row['cnpj'],
            'Faixa_CEP': faixa_correspondente['Faixa_CEP'].iloc[0],
            'Regiao': faixa_correspondente['Regiao'].iloc[0]
        }, ignore_index=True)
    else:
        print(f"Nenhuma faixa correspondente encontrada para o CEP {cep_atual}")

# Exibir o DataFrame resultante
resultado

resultado.to_csv('./Codigos_postais/Cnpj_Regiao', index=False)
```
